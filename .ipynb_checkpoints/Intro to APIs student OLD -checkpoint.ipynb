```python
url = "http://ergast.com/api/f1/2008/5/driverStandings.json"
```

```python
response = requests.get(url)   #get info from URL
print(response)
```

```python
#response object
help(response)
```

```python
#check and make sure we got a successful response from the API
response.ok
```

```python
#view the response as a JSON
response.json()    #gets dict from url in json which is a python dict
```

```python
type(response.json())
```

```python
# We only want the Driver Standings body part response.json gets whole thing in there then you want to index
#into mr data then standing table then standings list then index in pos 1 of list so [0] 
#pretty much eveytime a new tab starts 
data = response.json()["MRData"]["StandingsTable"]["StandingsLists"][0]["DriverStandings"]
data
```

```python
type(data)
```

```python
#get each racer by postion number
def get_racer_info(data):
    new_data=[]
    for racer in data:
        racer_dict={}
        racer_name = f"{racer['Driver']['givenName']} {racer['Driver']['familyName']}"
        
        racer_dict[racer_name]={
            'first_name':racer['Driver']['givenName'],
            'last_name':racer['Driver']['familyName'],
            'position':racer['position'],
            'wins':racer['wins'],
            'DOB':racer['Driver']['dateOfBirth'],
            'nationality':racer['Driver']['nationality'],
            'constructor':racer['Constructors'][0]['name'] #index into list of pos 0 then the names
        }
        new_data.append(racer_dict) #store racer_dict into new_data list and add.append
    return new_data
get_racer_info(data)
```

```python
def get_driver_info_by_year_and_round(year,r):   #you can add any year and round
    url = f"http://ergast.com/api/f1/{year}/{r}/driverStandings.json"
    response = requests.get(url)
    #if not response.ok:   #if response is bad and you cannot connect to database
    if response.json() ["MRData"]["StandingsTable"]["StandingsLists"]:
        return "WE had an error loading your data likely the year or round is not in the database"
    data = response.json()["MRData"]["StandingsTable"]["StandingsLists"][0]["DriverStandings"]
    return get_racer_info(data)

get_driver_info_by_year_and_round(1954,1)
```
